```python
def mix_valid():

	# le imagens de todos os datasets
	folder = './notebooks/train-folder/tmp/*/valid/images/*.jpg'
	all_train_images = glob( folder)

	# shuffle e seleciona 25 mil
	selected_train_imaegs= random.sample( all_train_images, 5000)

	# copia todas as imagens e os labels
	for i in range( 0, len( selected_train_imaegs)):
		if i%100 == 0:
			print( i)
		# image
		shutil.copyfile( selected_train_imaegs[i], 
			os.path.join( 
				'./notebooks/train-folder/tmp/mixed/valid/images', 
				Path(selected_train_imaegs[i]).name
			)
		)
		# label
		shutil.copyfile( 
			selected_train_imaegs[i].replace( 'images','labels').replace( 'jpg', 'pgn'),
			os.path.join( 
				'./notebooks/train-folder/tmp/mixed/valid/images', 
				Path(selected_train_imaegs[i]).name
			).replace( 'images','labels').replace( 'jpg', 'pgn')
		)   
		
mix_valid()
```
