```python
def read_all(paths):
    lines=[]
    for path in paths:
#         print( 'reading ', path)
        with open(path) as f:
            newlines = f.readlines()
#             print( path, len(newlines))
#             epochs.append( (len(newlines), path, newlines))
            lines.extend( newlines)
    return lines
```

```python
files = glob( os.path.join( 'C:/mestrado/repos-github/chess-attention/train-folder/log', 'train_icpr2022_comparativo_10k_sampled_*.txt'))
print( len( files))
```

```python
for f in files:
    print( Path( f).name)
```

```python
def fname2qtd( fname):
    print( fname)
    return fname.\
        replace('train_icpr2022_comparativo_10k_sampled_', '').\
        replace( '---8linhas-handwritten--10k--v20220408--nounk.txt', '')

def collect( f):
    lines = read_all([f])
    logs = [json.loads( line) for line in lines]
    
    r = {
        'train_loss': [log['train_loss'] for log in logs],
        'valid_loss': [log['valid_loss'] for log in logs],
        'train_acc': [log['train_acc'] for log in logs],
        'valid_acc': [log['valid_acc'] for log in logs],
        'test_acc': [log['test_acc'][0][0][-1] for log in logs],
        'test_acc0': [log['test_acc'][0][0][0] for log in logs],
    }
    return r

logs = []
for f in files:
    logs.append( (fname2qtd( Path( f).name), collect( f)))
    
logs = sorted( logs, key=lambda x: int( x[0]), reverse=True)
```

```python
for x in logs:
    print( x[0])
```

```python
figure(figsize=(10, 6))
legends = []
for qtd, log in logs:
    plt.plot( gaussian_filter1d(log['test_acc'], sigma=2))
    legends.append( "{}({:.2f}%)".format( qtd, log['test_acc'][-1]*100))
plt.ylabel( 'test acc')
plt.gca().legend( legends)
plt.show()
```
