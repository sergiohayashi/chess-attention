```python
#verifica que os arquivos do diretorio destinatario existe
def label_file_of_image( image_path):
    filename= Path( image_path).name
    return path.join( LABELS_PATH, filename+ '.pgn')
```

```python
def file_copy_if_exists( org, target):
    if not path.exists( org):
        return False
    print( org + " => "+ target)
    shutil.copy2( org, target)
    return True
```

```python
file_counter= 1
file_head= 'carnaval_original_'
#rotina para cada arquivo de image
def process_1file( org_image_file):
    global file_counter
    
    #1. gera novo nome
    new_file_name= file_head+ str( file_counter).zfill(4)
    file_counter= file_counter+1
    
    new_image_file= new_file_name+".jpg"
    new_label_file= new_file_name+".pgn"
    
    #3. copia o label de label_path para all_clean/lables
    exist= file_copy_if_exists(
        label_file_of_image( org_image_file),
        path.join( ALL_CLEAN_LABELS, new_label_file)
    )
        
    if exist:
        print( "copia imagem "+ org_image_file)
        #2. copia do diretorio de done, para all_clean/imges com novo nome
        file_copy_if_exists(
            org_image_file, 
            path.join( ALL_CLEAN_IMAGES, new_image_file)
        )
    else:
        print( "label nao encontrado para "+ org_image_file+". Descarta")
        #volta..
        file_counter= file_counter-1

    
    
```
