```python
def crop_only( img, y, h, x, w):
    crop_img= img[ y:y+h, x:x+w]
    return crop_img

def do_crop( filename):
    img= cv2.imread( filename)    
    y= 599
    h= 560
    x= 467
    w= 774
    crop_img= crop_only( img, 599, 560, 467, 774)
    final_img= np.concatenate((
        crop_only( crop_img, 0, 560, 0, 339), 
        crop_only( crop_img, 0, 560, 339+96, 339)
    ), axis=1)
    return final_img
```
