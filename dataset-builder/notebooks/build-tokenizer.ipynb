```python
def read( n):
    buf= []
    with open(file) as f:
        for x in range( n):
#             if( x%100== 0):
#                 clear_output(wait=True)
#                 print( x)
            buf.append( next( f))
    return buf[5:]

sample= read( 3000*1000)
```

```python
def filter_numeros( line):
    return re.sub( r"[W,B]\d+\.", '', line)
```

```python
def filter_chaves( s):
    _in= False
    r= []
    for x in s:
        if _in:
            if x== '}':
                _in= False
            continue

        #not in
        if x== '{':
            _in= True
            continue
            
        #not in & x!= '{'
        r.append( x)
    return ''.join(r)

filter_chaves( 'jdflkasjd{djklsafd} dfasdjf f{dkjfl}jk {}x)')
```

```python
def cleanup_lines( all_lines):
    filtered= [re.split( r'###', line)[1].lstrip().rstrip() for line in all_lines]
    filtered= [filter_chaves( line) for line in filtered]
    filtered= [filter_numeros( line) for line in filtered]
#     filtered= [line.split() for line in filtered]
#     all_words= np.concatenate( filtered)   # matriz para vetor..
#     unique, counts= np.unique( all_words, return_counts= True)
#     words= np.asarray( (unique, counts)).T
    return filtered

sample_lines= cleanup_lines( sample)
```

```python
def read_filtered_to_set( n):
    words= set()
    skip_lines= 5
    with open(file) as f:
        for _ in range( n):
#             if( x%100== 0):
#                 clear_output(wait=True)
#                 print( x)

            line= next( f)
#             print( line)
            if( skip_lines==0):
                line= re.split( r'###', line)[1].lstrip().rstrip()
                line= filter_chaves( line)
                line= filter_numeros( line)

                for w in line.split():
                    words.add( w)
            else:
                skip_lines-= 1

        return words

words= read_filtered_to_set( 30*1000*1000) #*1000)
print( len( words))
```

```python
def build_tokenizer_from_lines(labels):
#   labels=glob( '/content/carnaval_all_labels/carnaval_all_labels/labels/*.pgn')
#   labels= [read_label( f) for f in labels]
  labels= ['<start> ' + label + ' <end>' for label in labels]

  # Choose the top 5000 words from the vocabulary
  top_k = 10000
  tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=top_k,
                                                    oov_token="<unk>",
                                                    lower= False,
                                                    filters= ' ')
                                                    # filters='!"#$%&()*+.,-/:;=?@[\]^_`{|}~ ')

  #forca a usar sempre uma lista com todas as words com 1 ocorrecia de cada word
  tokenizer.fit_on_texts(labels)
  tokenizer.word_index['<pad>'] = 0
  tokenizer.index_word[0] = '<pad>'
  #tokenizer.fit_on_texts([" ".join( _unique_labels)])
  return tokenizer

tokenizer= build_tokenizer_from_lines( sample_lines)
```
