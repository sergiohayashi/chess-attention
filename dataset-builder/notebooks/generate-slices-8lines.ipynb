```python
def read_label( path):
  f= open( path)
  d= f.read()
  f.close()
  return d 

def write_label( path, text):
  f= open( path, "w")
  f.write( text)
  f.close()

def read_text_file( filename):
    f= open( filename, 'r')
    d= f.read()
    f.close()
    return d

def show( img):
    plt.imshow( img)
    plt.title( 'picture')
    plt.show()
    
    
def write_text_file( filename, data):
    f= open( filename, "w")
    f.write( data)
    f.close()
    #print( "Write text to "+ filename)


def show_image( img, name= 'noname'):
  plt.imshow( img)
  plt.title( name)
  plt.show()

def show_gimage( img, name= 'noname'):
  plt.imshow( img, cmap= 'gray')
  plt.title( name)
  plt.show()

def _load_image( path):
  # return cv2.imread( path, cv2.IMREAD_GRAYSCALE)
  return cv2.imread( path)
  
def write_image( path, img):
  cv2.imwrite( path, img)

def resize( img, scale_percent= 60, verbose=False):
  #scale_percent = 60 # percent of original size
  width = int(img.shape[1] * scale_percent / 100)
  height = int(img.shape[0] * scale_percent / 100)
  dim = (width, height)
  if verbose:
    print( "scale", scale_percent)
    print( "dim", dim)
  # resize image
  return cv2.resize(img, dim, interpolation = cv2.INTER_AREA) 
```

```python
source_dir= "C:\mestrado\datasets\carnaval_fullpage_clean"
img_files= glob( os.path.join( source_dir, 'images', "*.jpg"))
label_files= glob( os.path.join( source_dir, 'labels', "*.pgn"))
```

```python
print( len( img_files), len( label_files))
```

```python
H= 70
W= 338
W2= 769
spc= 97 #74
pad_left= 0
pad_top= 0
dy= 20

def crop_only( img, x, y, w, h):
#     print( y, y+h, x, x+w)
    crop_img= img[ y:y+h+dy, x:x+w]
    return crop_img

def crop( img, ln, col):
    x= pad_left if col== 1 else pad_left+2*W+spc
    y= pad_top + ln*H
    img= crop_only( img, x, y, W2, 8*H)
    return np.concatenate((
        crop_only( img, 0, 0, W, 8*H), 
        crop_only( img, W+spc, 0, W, 8*H)
    ), axis=1)
    return img


folder= "C:\mestrado\datasets\carnaval_8lines_subseq"

def generate_for( idx):
#     print( img_files[idx])

    img= cv2.imread( img_files[idx])
    label= read_text_file( label_files[idx])

#     show( img)

    n= len( label.split())
    n_left= int(min( 25-7, n/2-7))
    n_right= int(min( 25-7, n/2 - 25 - 7))
    
#     print( n_left, n_right)

    labels= label.split()

    count= 0
    for ln in range( 0, n_left):
#         print( '----left - ', ln, '-----')
       
        new_img= crop( img, ln, 1)
        tokens= labels[ln*2:ln*2+16]
        new_pgn= " ".join( tokens)
        filename= Path( img_files[idx]).name
        new_name= tokens[0].ljust( 5, '_')+"_"+tokens[1].ljust( 5, '_')+"_"+filename[9:13]+ "_"+ str(count).zfill( 4)+"L"
        new_image_file= path.join( folder, "images", new_name+ '.jpg')
        new_label_file= path.join( folder, "labels", new_name+ '.pgn')

#         show( new_img)
#         print( tokens)
        cv2.imwrite( new_image_file, new_img)
        write_text_file( new_label_file, new_pgn)
        count= count+1
    
    for ln in range( 0, n_right):
#         print( '---right-', ln, '-----')
       
        new_img= crop( img, ln, 2)
        tokens= labels[(25+ln)*2:(25+ln)*2+16]
#         tokens= labels[ln*2:ln*2+16]
        new_pgn= " ".join( tokens)
        filename= Path( img_files[idx]).name
        new_name= tokens[0].ljust( 5, '_')+"_"+tokens[1].ljust( 5, '_')+"_"+filename[9:13]+ "_"+ str(count).zfill( 4)+"R"
        new_image_file= path.join( folder, "images", new_name+ '.jpg')
        new_label_file= path.join( folder, "labels", new_name+ '.pgn')

#         show( new_img)
#         print( tokens)

        cv2.imwrite( new_image_file, new_img)
        write_text_file( new_label_file, new_pgn)
        count= count+1
    

for i in range( 0, len( img_files)):
    print( 'generating for ', i, '....')
    generate_for( i)
```
